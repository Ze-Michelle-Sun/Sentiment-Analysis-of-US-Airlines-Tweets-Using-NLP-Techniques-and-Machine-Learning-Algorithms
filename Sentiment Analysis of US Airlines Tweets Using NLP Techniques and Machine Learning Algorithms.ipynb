Now we need to check the number of words use for the positive and negative sentiments, So we will use wordcloud for that.

### Wordcloud for Positive Reason

```python
# Cleaning the text by removing punctuations and converting to lowercase
ens_learn_temp =df

ens_learn_temp['text'] = ens_learn_temp['text'].apply(lambda x: re.sub('[,\.!?]', '', x))
ens_learn_temp['text'] = ens_learn_temp['text'].apply(lambda x: x.lower())

# Filtering positive sentiment tweets and joining their words
pos_cloud = ens_learn_temp[ens_learn_temp['airline_sentiment']=='positive']
overall_words_pos = ' '.join(pos_cloud['text'])

# Cleaning the words by removing urls, usernames, and retweets
cleaned_word_pos = " ".join([word for word in overall_words_pos.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and word != 'RT'
                            ])

# Generating a word cloud for positive sentiment tweets
word_cloud_pos = WordCloud(stopwords=STOPWORDS,
                      background_color='White',
                      width=2500,
                      height=1800
                     ).generate(cleaned_word_pos)
plt.figure(1,figsize=(8, 8))
plt.imshow(word_cloud_pos,interpolation='bilinear')
plt.axis('off')
plt.title('Most Used Positive Words', fontsize=25)
plt.show()
```

### Wordcloud for Negative Reason

```python
# First, we create a copy of the dataframe
ens_learn_temp = df.copy()

# Remove punctuations and convert all text to lowercase
ens_learn_temp['text'] = ens_learn_temp['text'].map(lambda x: re.sub('[,.!?]', '', x))
ens_learn_temp['text'] = ens_learn_temp['text'].map(lambda x: x.lower())

# Select only the negative tweets
neg_cloud = ens_learn_temp[ens_learn_temp['airline_sentiment'] == 'negative']

# Join all the negative tweets into one string
overall_words_neg = ' '.join(neg_cloud['text'])

# Clean the string by removing urls, mentions, and RT
cleaned_word_neg = " ".join([word for word in overall_words_neg.split()
    if 'http' not in word
    and not word.startswith('@')
    and word != 'RT'])

# Generate a word cloud with the cleaned words
word_cloud_neg = WordCloud(stopwords=STOPWORDS,
background_color='white',
width=2500,
height=1800
).generate(cleaned_word_neg)

# Plot the word cloud
plt.figure(figsize=(8, 8))
plt.imshow(word_cloud_neg, interpolation='bilinear')
plt.axis('off')
plt.title('Most Used Negative Words', fontsize=25)
plt.show()
```

```python
# Clean tweet text
def clean_tweet_text(tweet):
    # Remove @ mentions
    tweet = re.sub(r'@\w+', '', tweet)
    
    # Remove URLs
    url = re.compile(r'https?://\S+|www\.\S+')
    tweet = url.sub(r'', tweet)
    
    # Remove HTML tags
    html = re.compile(r'<.*?>')
    tweet = html.sub(r'', tweet)
    
    # Remove emojis
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  
                           u"\U0001F300-\U0001F5FF" 
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    tweet = emoji_pattern.sub(r'', tweet)
    
    # Fix text
    tweet = re.sub(r"Mr\'s", " he is", tweet)
    tweet = re.sub(r"Mr\'.", " he is", tweet)
    tweet = re.sub(r"Ms\'.", " She ", tweet)
    tweet = re.sub(r"She\'s", " she is", tweet)
    tweet = re.sub(r"I\'m", " i am", tweet)
    tweet = re.sub(r"haven\'t", " have not", tweet)
    tweet = re.sub(r"you\'r", " you are", tweet)
    tweet = re.sub(r"won\'t", " will not", tweet)
    tweet = re.sub(r"won\'t've", " will not have", tweet)
    tweet = re.sub(r"can\'t", " can not", tweet)
    tweet = re.sub(r"don\'t", " do not", tweet)
    tweet = re.sub(r" plz", " please", tweet)
    tweet = re.sub(r"it\'s", " its", tweet) 
    tweet = re.sub(r"can\'t've", " can not have", tweet)
    tweet = re.sub(r"ma\'am", " madam", tweet)
    tweet = re.sub(r"let\'s", " let us", tweet)
    tweet = re.sub(r"ain\'t", " am not", tweet)
    tweet = re.sub(r"shan\'t", " shall not", tweet)
    tweet = re.sub(r"sha\n't", " shall not", tweet)
    tweet = re.sub(r"o\'clock", " of the clock", tweet)
    tweet = re.sub(r"y\'all", " you all", tweet)
    tweet = re.sub(r"n\'t", " not", tweet)
    tweet = re.sub(r"n\'t've", " not have", tweet)
    tweet = re.sub(r"\'re", " are", tweet)
    tweet = re.sub(r"\'s", " is", tweet)
    tweet = re.sub(r"\'d", " would", tweet)
    tweet = re.sub(r"\'d've", " would have", tweet)
    tweet = re.sub(r"\'ll", " will", tweet)
    tweet = re.sub(r"\'ll've", " will have", tweet)
    tweet = re.sub(r"\'t", " not", tweet)
    tweet = re.sub(r"\'ve", " have", tweet)
    tweet = re.sub(r"\'m", " am", tweet)
    tweet = re.sub(r"\'re", " are", tweet)
    tweet = re.sub(r"He\'s", " he is", tweet)
    tweet = re.sub(r"in\'s", " in", tweet)
    
    # Separate alphabets and digits
    alphanumeric_pattern = re.compile(r"[^\W\d_]+|\d+")
    tweet = " ".join(alphanumeric_pattern.findall(tweet))
    
    # Remove non-alphabetic characters and convert to lowercase
    tweet = re.sub("[^a-zA-Z]", " ", tweet).lower()
    
    # Replace "plz" with "please"
    tweet = re.sub(r"\bplz\b", "please", tweet)
    
    return tweet
```

# <font color='orange'>Explanation of Text-Representations in NLP</font>

Before moving forward in sentimental analysis, we will first see some of the text-representation techniques of NLP. In Natural Language Processing (NLP) the conversion of raw-text to numerical form is called Text Representation.

## Processing Text of Tweets

```python
def process_text(text):
    # Remove special characters and punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Convert to lowercase and split into words
    words = text.lower().split()
    return words

df['words'] = df['final_text'].apply(process_text)
```

**There are several techniques for representing text data:**

**One-Hot Encoding:** This technique represents categorical data as binary vectors. In the context of text data, one-hot encoding is often used to represent individual words or entire documents.

**Bag-of-Words (BoW):** This technique represents each document as a vector of word counts, where the position of each word corresponds to its index in the vocabulary.

**Term Frequency-Inverse Document Frequency (TF-IDF):** This technique represents each document as a vector of term frequencies weighted by the inverse document frequency. It assigns higher weights to words that are more important and less frequent in the corpus.

**Word2Vec:** This is a popular algorithm used to create text representations in NLP. It is a neural network-based approach that learns word embeddings by predicting a word's context or its neighboring words.

## One-Hot Encoding

```python
# vocabulary of words present in dataset
df_vocab = []
for text in df.words:
    for word in text:
        if word not in df_vocab:
            df_vocab.append(word)
```

```python
#function to return one-hot representation of passed text
def get_onehot_representation(text, vocab = df_vocab):
    onehot_encoded = []
    for word in text:
        temp = [0]*len(vocab)
        temp[vocab.index(word)-1] = 1
        onehot_encoded.append(temp)
    return onehot_encoded
```
